```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def getWindyMDP():
    # set of states - 6x8 grid:
    S = []
    for x in range(0,6):
        for y in range(0,8):
            S.append([x, y])

    # set of terminal states:
    Sterminal = [[3,5]]

    # set of actions:
    # [left, up, right, down]
    A = [[0, -1], [-1, 0],[0, 1], [1, 0]]

    # auxiliary wind vector:
    wind = [0, 0, 1, 1, 2, 2, 0, 0]

    P = np.zeros((len(A), len(S), len(S)))
    # state transition probability tensor P^a_s,s'
    # P[action, from, to]
    for s, state in enumerate(S):
        for a, action in enumerate(A):
            nextState = [0,0]
            nextState[0] = state[0] + action[0]
            nextState[1] = state[1] + action[1]
            nextState[1] = np.clip(nextState[1], 0, 7)
            nextState[0] -= wind[nextState[1]]
            nextState[0] = np.clip(nextState[0], 0, 5)
            sPrime = S.index(nextState)
            P[a, s, sPrime] = 1

    R = np.ones((len(S), len(A))) * -1.0
    rewardedActions = np.where(P[:,:,3*8+5] > 0)
    for i, action in enumerate(rewardedActions[0]):
        R[rewardedActions[1][i], action] = 10
#     R[S.index([2,5]), 3] = 10
#     R[S.index([3,4]), 2] = 10
#     R[S.index([4,5]), 1] = 10
#     R[S.index([5,6]), 0] = 10

    gamma = 0.9

    return {'S': S, 'Sterminal': Sterminal, 'A' : A, 'P' : P, 'R' : R, 'gamma' : gamma, 'wind' : wind}
```

```python
def getUniformRandomPolicy(S, A):
    policy = np.ones((len(S), len(A))) * (1.0 / len(A))
    return policy
```

```python
def getPOverPolicy(P, policy):
    Ppolicy = np.zeros(P.shape[1:])
    for a in range(0, P.shape[0]):
        Ppolicy += P[a,:,:] * policy[:,a]
    return Ppolicy

def getROverPolicy(R, policy):
    Rpolicy = np.zeros(R.shape[0])
    for a in range(0, R.shape[1]):
        Rpolicy += R[:,a] * policy[:, a]
    return Rpolicy

def getAnalyticSolution(mdp, policy):
    Ppolicy = getPOverPolicy(mdp['P'], policy)
    Rpolicy = getROverPolicy(mdp['R'], policy)
    v = np.linalg.inv(np.identity(len(mdp['S'])) - mdp['gamma'] * Ppolicy).dot(Rpolicy)
    return v

mdp = getWindyMDP()
policy = getUniformRandomPolicy(mdp['S'], mdp['A'])
v = getAnalyticSolution(mdp, policy)
```

```python
def drawStateValues(v, policy, nX=6, nY=8, title=None):
    vMap = np.reshape(v, (nX, nY))
    plt.imshow(vMap, cmap='RdYlGn', interpolation='nearest')
    plt.colorbar()
    if title != None:
        plt.title(title)
    ax = plt.axes()
    plt.xticks(range(0,8), mdp['wind'])
    ax.get_yaxis().set_ticks([])
    for i in range(0, policy.shape[0]):
        x = mdp['S'][i][1]
        y = mdp['S'][i][0]
        dLeft = policy[i][0] * 0.25
        dUp = policy[i][1] * 0.25
        dRight = policy[i][2] * 0.25
        dDown = policy[i][3] * 0.25
        if dLeft > 0:
            ax.arrow(x, y, -dLeft, 0, head_width=0.2, head_length=0.2, fc='white', ec='grey')
        if dUp > 0:
            ax.arrow(x, y, 0, -dUp, head_width=0.2, head_length=0.2, fc='white', ec='grey')
        if dRight > 0:
            ax.arrow(x, y, dRight, 0, head_width=0.2, head_length=0.2, fc='white', ec='grey')
        if dDown > 0:
            ax.arrow(x, y, 0, dDown, head_width=0.2, head_length=0.2, fc='white', ec='grey')
    plt.show()

mdp = getWindyMDP()
policy = getUniformRandomPolicy(mdp['S'], mdp['A'])
v = getAnalyticSolution(mdp, policy)
drawStateValues(v, policy, title='Values with uniform random policy')
```

```python
def policyIteration(mdp, iterations=2):
    policy = getUniformRandomPolicy(mdp['S'], mdp['A'])
    while True:
        v = getAnalyticSolution(mdp, policy)
        oldPolicy = policy
        policy = np.zeros((policy.shape))
        # choose optimal actions for new policy:
        bestActions = np.argmax(mdp['R'].T + mdp['gamma'] * np.sum(mdp['P'] * v, axis = 2), axis=0)
        policy[np.arange(0, len(mdp['S'])), bestActions] = 1
        if (policy == policy).all():
            break
    return policy

mdp = getWindyMDP()
policy = policyIteration(mdp)
v = getAnalyticSolution(mdp, policy)
drawStateValues(v, policy, title='Values with policy iteration')
```

```python
def drawPath(mdp, policy, startpoint):
    currentIndex = startpoint[0] * 8 + startpoint[1]
    walkingXs = []
    walkingYs = []
    stepCounter = 0
    while True:
        walkingYs.append(mdp['S'][currentIndex][0])
        walkingXs.append(mdp['S'][currentIndex][1])
        if mdp['S'][currentIndex] == mdp['Sterminal'][0]:
            break
        action = np.argmax(policy[currentIndex])
        currentIndex = np.argmax(mdp['P'][action, currentIndex, :])
        stepCounter += 1
    vMap = np.reshape(v, (6, 8))
    plt.imshow(vMap, cmap='RdYlGn', interpolation='nearest')
    plt.colorbar()
    ax = plt.axes()
    plt.xticks(range(0,8), mdp['wind'])
    ax.get_yaxis().set_ticks([])
    plt.scatter([5],[3], color='white')
    plt.scatter(startpoint[1], startpoint[0], color='white')
    plt.plot(walkingXs, walkingYs, color='white')
    plt.title('Walk from ' + str(startpoint) + ' in ' + str(stepCounter) + ' steps')
    plt.show()
    
    

mdp = getWindyMDP()
policy = policyIteration(mdp)
startpoint = [5,0]
drawPath(mdp, policy, startpoint)
```

```python
def getSampledEpisode(mdp, policy):
    currentState = np.random.randint(0, len(mdp['S']))
    sampledStates = []
    sampledActions = []
    sampledRewards = []
    while True:
        if mdp['S'][currentState] == mdp['Sterminal'][0]:
            break
        sampledStates.append(currentState)
        action = np.random.choice(4, p=policy[currentState])
        sampledActions.append(action)
        sampledRewards.append(mdp['R'][currentState, action])
        currentState = np.argmax(mdp['P'][action, currentState, :])
        
    return sampledStates, sampledActions, sampledRewards

mdp = getWindyMDP()
policy = getUniformRandomPolicy(mdp['S'], mdp['A'])
episode = getSampledEpisode(mdp, policy)
print('sampled episode of length ' + str(len(episode[0])))
print('last ten states, actions & rewards:')
print(episode[0][-10:])
print(episode[1][-10:])
print(episode[2][-10:])
print('overall reward: ' + str(np.sum(episode[2])))
```

```python
def policyEvalMC(mdp, policy, iterations=1000):
    Ns = np.zeros(len(mdp['S']))
    Ss = np.zeros(len(mdp['S']))
    for i in range(0, iterations):
        sampS, sampA, sampR = getSampledEpisode(mdp, policy)
        Gt = 0
        for j in range(1,len(sampS) + 1):
            Gt = sampR[-j] + mdp['gamma'] * Gt
            s = sampS[-j]
            Ns[s] += 1
            Ss[s] += Gt
    vs = Ss
    vs[np.where(Ns > 0)] /= Ns[np.where(Ns > 0)]
    vs[np.where(Ns == 0)] = float('nan')
    return vs
```

```python
mdp = getWindyMDP()
policy = policyIteration(mdp)
v = policyEvalMC(mdp, policy)
drawStateValues(v, policy, title='Optimal policy, evaluated with MC')
```

```python
def policyEvalTDlambda(mdp, policy, lambd=0.5, alpha=0.1, iterations=1000):
    Vs = np.zeros(len(mdp['S']))
    for i in range(0, iterations):
        sampS, sampA, sampR = getSampledEpisode(mdp, policy)
        Es = np.zeros(len(mdp['S']))
        # explicitly state that terminal state is last:
        sampS.append(3*8+5)
        for j in range(0,len(sampS)-1):
            currentState = sampS[j]
            Es = mdp['gamma'] * lambd * Es
            Es[currentState] += 1
            nextState = sampS[j+1]
            delta = sampR[j] + mdp['gamma'] * Vs[nextState] - Vs[currentState]
            Vs = Vs + alpha * delta * Es
    Vs[3*8+5] = float('nan')
    return Vs
```

```python
currentState = 5
mdp = getWindyMDP()
policy = policyIteration(mdp)
v = policyEvalTDlambda(mdp, policy, lambd=0.5)
drawStateValues(v, policy, title='Optimal policy, evaluated with TD(0.5)')
```
